```python
import pygame
import sys

pygame.init()

WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("play one")
background = pygame.image.load('fon.jpg')
background_width = background.get_width()


speed = 5
bg_x = 0

clock = pygame.time.Clock()

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    keys = pygame.key.get_pressed()
    if keys[pygame.K_RIGHT]:
        bg_x -= speed 
    elif keys[pygame.K_LEFT]:
        bg_x += speed 

    screen.fill((0, 0, 0)) 
    screen.blit(background, (bg_x, 0))
    
    if bg_x <= -background_width:
        bg_x = 0

    pygame.display.flip()  
    clock.tick(60)  
```

```python
import pygame
import time

pygame.init()

window_width = 800
window_height = 600
window = pygame.display.set_mode((window_width, window_height))
pygame.display.set_caption("Игровое окно")

fon = 'fon.jpg' 
img1 = pygame.image.load(fon)
back_fon = pygame.transform.scale(img1, (window_width, window_height)) 

speed = 1  
sdvig_fona = 0 

class Player(pygame.sprite.Sprite):
    def __init__(self, filename, hero_x=100, hero_y=250, x_speed=0, y_speed=0):
        super().__init__()
        self.image = pygame.image.load(filename)
        self.rect = self.image.get_rect()
        self.hero_x = hero_x
        self.hero_y = hero_y
        self.rect.x = self.hero_x
        self.rect.y = self.hero_y
        self.x_speed = x_speed
        self.y_speed = y_speed

    def update(self):
        '''Перемещает персонажа, применяя текущую горизонтальную и вертикальную скорость'''
        self.rect.x += self.x_speed
        self.rect.y += self.y_speed

filename = 'apple.png'  
hero = Player(filename)
all_sprites = pygame.sprite.Group()
all_sprites.add(hero)

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                hero.x_speed = -5  # Влево
            if event.key == pygame.K_RIGHT:
                hero.x_speed = 5  # Вправо
            if event.key == pygame.K_UP:
                hero.y_speed = -5  # Вверх
            if event.key == pygame.K_DOWN:
                hero.y_speed = 5  # Вниз
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                hero.x_speed = 0
            if event.key == pygame.K_UP or event.key == pygame.K_DOWN:
                hero.y_speed = 0

    keys = pygame.key.get_pressed()
    if keys[pygame.K_RIGHT]:
        sdvig_fona -= speed  
    elif keys[pygame.K_LEFT]:
        sdvig_fona += speed  

    window.fill((0, 0, 0))  
    window.blit(back_fon, (sdvig_fona, 0))  
    
    if sdvig_fona <= -window_width:
        sdvig_fona = 0

   
    all_sprites.update()
    all_sprites.draw(window)

    pygame.display.update()  
    time.sleep(0.02)  
    
pygame.quit()
```

```python
pygame.init()


WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Игра с коллизиями")


WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)
GREEN = (0, 255, 0)

class Hero(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.Surface((50, 50))
        self.image.fill(BLUE)
        self.rect = self.image.get_rect()
        self.rect.center = (WIDTH // 2, HEIGHT - 60)
        self.speed = 5

    def update(self):
        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT]:
            self.rect.x -= self.speed
        if keys[pygame.K_RIGHT]:
            self.rect.x += self.speed
        if keys[pygame.K_UP]:
            self.rect.y -= self.speed
        if keys[pygame.K_DOWN]:
            self.rect.y += self.speed

class Enemy(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.Surface((50, 50))
        self.image.fill(RED)
        self.rect = self.image.get_rect()
        self.rect.x = random.randint(0, WIDTH - 50)
        self.rect.y = random.randint(0, HEIGHT // 2)
        self.speed = 2

    def update(self):
        self.rect.y += self.speed
        if self.rect.top > HEIGHT:
            self.rect.bottom = 0
            self.rect.x = random.randint(0, WIDTH - 50)

class Arrow(pygame.sprite.Sprite):
    def __init__(self, hero):
        super().__init__()
        self.image = pygame.Surface((5, 10))
        self.image.fill(GREEN)
        self.rect = self.image.get_rect()
        self.rect.center = hero.rect.center
        self.speed = 7

    def update(self):
        self.rect.y -= self.speed
        if self.rect.bottom < 0:
            self.kill()  

def main():
    hero = Hero()
    hero2 = Enemy()
    hero3 = Enemy()
    all_sprites = pygame.sprite.Group()
    all_sprites.add(hero, hero2, hero3)

    arrows = pygame.sprite.Group()

    running = True
    clock = pygame.time.Clock()

    while running:
        screen.fill(WHITE)

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:
                    arrow = Arrow(hero)
                    all_sprites.add(arrow)
                    arrows.add(arrow)

    
        all_sprites.update()

        for arrow in arrows:
            enemy_hit = pygame.sprite.spritecollide(arrow, all_sprites, False)
            for enemy in enemy_hit:
                if isinstance(enemy, Enemy):
                    arrow.kill()
                    enemy.kill()

        all_sprites.draw(screen)

        pygame.display.flip()
        clock.tick(60)

    pygame.quit()

if __name__ == "__main__":
    main()
```
