```python
pygame.init()

WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Анимация графических фигур")

clock = pygame.time.Clock()

def random_color():
    return (random.randint(0,255), random.randint(0,255), random.randint(0,255))

class MovingShape:
    def __init__(self, shape_type, rect, color, speed):
        self.shape_type = shape_type 
        self.rect = pygame.Rect(rect) 
        self.color = color
        self.speed = speed  

        if self.shape_type == 'circle':
            self.radius = self.rect.width // 2

    def move(self):
        self.rect.x += self.speed
        # Отскок от границ
        if self.rect.right >= WIDTH or self.rect.left <= 0:
            self.speed = -self.speed
            self.color = random_color()

    def draw(self, surface):
        if self.shape_type == 'square' or self.shape_type == 'rect':
            pygame.draw.rect(surface, self.color, self.rect)
        elif self.shape_type == 'circle':
            center = (self.rect.x + self.radius, self.rect.y + self.radius)
            pygame.draw.circle(surface, self.color, center, self.radius)
        elif self.shape_type == 'triangle':
            point1 = (self.rect.centerx, self.rect.top)
            point2 = (self.rect.left, self.rect.bottom)
            point3 = (self.rect.right, self.rect.bottom)
            pygame.draw.polygon(surface, self.color, [point1, point2, point3])

    def is_clicked(self, pos):
        if self.shape_type in ['square', 'rect', 'triangle']:
            return self.rect.collidepoint(pos)
        elif self.shape_type == 'circle':
            center = (self.rect.x + self.radius, self.rect.y + self.radius)
            distance = ((pos[0]-center[0])**2 + (pos[1]-center[1])**2)**0.5
            return distance <= self.radius

shapes = [
    MovingShape('square', (50, 100, 50, 50), random_color(), 5),
    MovingShape('rect', (150, 200, 100, 50), random_color(), 3),
    MovingShape('circle', (300, 300, 60, 60), random_color(), 4),
    MovingShape('triangle', (500, 100, 80, 80), random_color(), 6)
]

running = True
while running:
    clock.tick(60)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN:
            pos = pygame.mouse.get_pos()
            for shape in shapes:
                if shape.is_clicked(pos):
                    shape.color = random_color()

    screen.fill(WHITE)

    for shape in shapes:
        shape.move()
        shape.draw(screen)

    pygame.display.flip()

pygame.quit()
```

```python
pygame.init()

WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLACK = (0, 0, 0)
YELLOW = (255, 255, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)

def draw_task_1():
    screen = pygame.display.set_mode((800, 600), pygame.RESIZABLE)
    pygame.display.set_caption("Задание 1: Красный закрашенный круг")

    screen.fill(WHITE)

    pygame.draw.circle(screen, RED, (100, 100), 50)
    pygame.display.flip()

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
    pygame.quit()

def draw_task_2():
    screen = pygame.display.set_mode((800, 600), pygame.RESIZABLE)
    pygame.display.set_caption("Задание 2: Круг с контуром")

    screen.fill(WHITE)
    pygame.draw.circle(screen, BLACK, (500, 500), 40, 15)
    pygame.display.flip()

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
    pygame.quit()

def draw_task_3():
    screen = pygame.display.set_mode((800, 600), pygame.RESIZABLE)
    pygame.display.set_caption("Задание 3: Круг в центре экрана")

    screen.fill(WHITE)
    
    center = (800 // 2, 600 // 2)
    pygame.draw.circle(screen, GREEN, center, 100)
    pygame.display.flip()

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
    pygame.quit()

def draw_task_4():
    screen = pygame.display.set_mode((800, 600), pygame.RESIZABLE)
    pygame.display.set_caption("Задание 4: Прямоугольник 300x200, закрашенный")

    screen.fill(WHITE)
    pygame.draw.rect(screen, YELLOW, [50, 50, 300, 200])
    pygame.display.flip()

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
    pygame.quit()

def draw_task_5():
    screen = pygame.display.set_mode((800, 600), pygame.RESIZABLE)
    pygame.display.set_caption("Задание 5: Пять случайных прямоугольников")

    screen.fill(WHITE)

    for _ in range(5):
        width = random.randint(20, 100)
        height = random.randint(20, 100)
        x = random.randint(0, 800 - width)
        y = random.randint(0, 600 - height)
        color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
        pygame.draw.rect(screen, color, [x, y, width, height])

    pygame.display.flip()

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
    pygame.quit()

def draw_task_6():
    screen = pygame.display.set_mode((800, 600), pygame.RESIZABLE)
    pygame.display.set_caption("Задание 6: Домик из линий")
    screen.fill(WHITE)

    def draw_house(surface):
        roof_top = (400, 200)
        left_roof = (300, 300)
        right_roof = (500, 300)

        pygame.draw.polygon(surface, (200, 50, 50), [roof_top, left_roof, right_roof])
        pygame.draw.rect(surface, (150, 200, 150), (300, 300, 200, 150))

    draw_house(screen)

    pygame.display.flip()
    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
    pygame.quit()

def draw_task_7():
    screen = pygame.display.set_mode((800, 600), pygame.RESIZABLE)
    pygame.display.set_caption("Задание 7: Произвольная фигура из линий")

    screen.fill(WHITE)

    points = [(random.randint(0, 800), random.randint(0, 600)) for _ in range(5)]
    pygame.draw.lines(screen, BLUE, True, points, 3)

    pygame.display.flip()

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
    pygame.quit()

def draw_task_8():
    screen = pygame.display.set_mode((800, 600), pygame.RESIZABLE)
    pygame.display.set_caption("Задание 8: Вывод изображения")

    screen.fill(WHITE)

    apple = pygame.image.load("apple.png")  
    screen.blit(apple, (600, 400)) 
    pygame.display.flip()

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
    pygame.quit()

def draw_task_9():
    screen = pygame.display.set_mode((800, 600), pygame.RESIZABLE)
    pygame.display.set_caption("Задание 9: Перемещение изображения")

    screen.fill(WHITE)

    apple = pygame.image.load("apple.png") 
    screen.blit(apple, (600, 400)) 

    pygame.draw.rect(screen, WHITE, [600, 400, 100, 100]) 
    screen.blit(apple, (200, 200))  # Новое место для яблока
    
    pygame.display.flip()

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
    pygame.quit()
```
